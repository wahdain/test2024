# Working with MultiIndex DataFrame

```python
import pandas as pd
import numpy as np

# Create a sample MultiIndex DataFrame
dates = pd.date_range('2023-01-01', periods=5)
tickers = ['AAPL', 'MSFT', 'GOOGL']
columns = pd.MultiIndex.from_product([tickers, ['Open', 'High', 'Low', 'Close', 'Volume', 'RSI']])
data = np.random.randn(5, 18)

print(dates)
print(tickers)
print(columns)
print(data)
```

```python
multi_index_df = pd.DataFrame(data, index=dates, columns=columns)
multi_index_df
```

```python
#Accessing Data

#Selecting Columns (.xs (cross-section))

# Select all 'Close' prices for all tickers
close_prices = multi_index_df.xs('Close', level=1, axis=1)
print(close_prices)

# Select 'Open' and 'Close' prices for 'AAPL'
aapl_prices = multi_index_df['AAPL'][['Open', 'Close']]
print(aapl_prices)

# Select all columns for 'AAPL'
aapl_all = multi_index_df['AAPL']
print(aapl_all)
```

```python
# Selecting Rows

# Select data for a specific date
date_data = multi_index_df.loc['2023-01-01']
print(date_data)

# Select data for a range of dates
date_range_data = multi_index_df.loc['2023-01-01':'2023-01-03']
print(date_range_data)
```

```python
# Advanced Selection

# Select 'Close' prices for 'AAPL' on specific dates
aapl_close_selected_dates = multi_index_df.loc['2023-01-01':'2023-01-03', ('AAPL', 'Close')]
print(aapl_close_selected_dates)

# Select 'Close' and 'RSI' for all tickers on a specific date
selected_columns = multi_index_df.loc['2023-01-01', (slice(None), ['Close', 'RSI'])]
print(selected_columns)
```

```python
# Modifying Data

# Assigning New Values

# Assign a new value to a specific cell
multi_index_df.loc['2023-01-01', ('AAPL', 'Close')] = 150.0

# Assign new values to a specific column
multi_index_df.loc[:, ('AAPL', 'RSI')] = 50.0

# Assign new values to a specific row
multi_index_df.loc['2023-01-01'] = np.random.randn(18)
```

```python
# Adding New Data

# Add a new date
new_data = pd.DataFrame(np.random.randn(1, 18), index=[pd.Timestamp('2023-01-06')], columns=columns)
multi_index_df = pd.concat([multi_index_df, new_data])

# Add a new ticker
new_ticker_data = pd.DataFrame(np.random.randn(5, 6), index=dates, columns=pd.MultiIndex.from_product([['NFLX'], ['Open', 'High', 'Low', 'Close', 'Volume', 'RSI']]))
multi_index_df = pd.concat([multi_index_df, new_ticker_data], axis=1)
```

```python
pd.set_option.
multi_index_df
```

```python
# Dropping Data

# Dropping Columns

# Drop a specific column
multi_index_df = multi_index_df.drop(columns=('AAPL', 'RSI'))

# Drop multiple columns
multi_index_df = multi_index_df.drop(columns=[('AAPL', 'RSI'), ('MSFT', 'Volume')])
```

```python
# Dropping Rows

# Drop a specific row
multi_index_df = multi_index_df.drop(index=pd.Timestamp('2023-01-01'))

# Drop multiple rows
multi_index_df = multi_index_df.drop(index=[pd.Timestamp('2023-01-01'), pd.Timestamp('2023-01-02')])
```

```python
# Resetting the MultiIndex

# Reset the column MultiIndex
flat_columns_df = multi_index_df.copy()
flat_columns_df.columns = ['_'.join(col).strip() for col in flat_columns_df.columns.values]
print(flat_columns_df)

# Reset the row index
reset_index_df = multi_index_df.reset_index()
print(reset_index_df)
```
